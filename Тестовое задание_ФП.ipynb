```python
import pandas as pd

audience = pd.read_csv('Данные для тестового задания - Данные об аудитории.csv')

audience.head()
```

```python
audience.info()
```

```python
audience['date'] = pd.to_datetime(audience['date'])
```

```python
mau = audience['user_id'].nunique()

mau
```

```python
dau_by_day = audience.groupby('date')['user_id'].nunique()

dau_by_day
```

```python
avg_dau = dau_by_day.mean()

avg_dau
```

```python
nov_1_users = audience[audience['date'] == '2023-11-01']['user_id'].unique()

nov_2_users = audience[audience['date'] == '2023-11-02']['user_id'].unique()

returned_users = len(set(nov_1_users) & set(nov_2_users))

retention_day_1 = returned_users / len(nov_1_users) * 100

retention_day_1
```

```python
users_with_adverts = audience[audience['view_adverts'] > 0]['user_id'].nunique()

conversion_to_adverts = users_with_adverts / mau * 100

conversion_to_adverts
```

```python
total_adverts = audience['view_adverts'].sum()

avg_adverts_per_user = total_adverts / mau

avg_adverts_per_user
```

```python
ab_tests = pd.read_csv('Данные для тестового задания - Данные АБ тестов.csv')

ab_tests.head()
```

```python
ab_tests.info()
```

```python
ab_tests['experiment_group'].unique()
```

```python
from scipy.stats import ttest_ind

for exp in sorted(ab_tests['experiment_num'].unique()):
    test_data = ab_tests[ab_tests['experiment_num'] == exp]
    
    group_control = test_data[test_data['experiment_group'] == 'control']['revenue']
    group_test = test_data[test_data['experiment_group'] == 'test']['revenue']
    
    stat, p_value = ttest_ind(group_control, group_test, equal_var=False)
    
    print('Эксперимент:', exp)
    print('ARPU control:', group_control.mean())
    print('ARPU test:', group_test.mean())
    print('p-value:', p_value)
    print()
```

```python
from statsmodels.stats.proportion import proportions_ztest

payments = [1003, 1099]
visitors = [100047501, 100001055]

stat, p_value = proportions_ztest(payments, visitors)

conversion_a = payments[0] / visitors[0] * 100
conversion_b = payments[1] / visitors[1] * 100

print('Конверсия A:', conversion_a)
print('Конверсия B:', conversion_b)
print('p-value:', p_value)
```
